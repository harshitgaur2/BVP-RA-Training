```python
#sigmoid(0, 1)
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

inp = np.linspace(-10, 10, 100)

def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
plt.plot(inp, sigmoid(inp), c='r')
```

```python
#tanh(-1, 1)
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x)) 
```

```python
plt.plot(inp, tanh(inp), c='b')
```

```python
#relu(0, inf)
def relu(x):
  return max(0, x)

p = pd.Series(inp)
q = p.apply(relu)
```

```python
plt.plot(p, q, c='g')
```

```python
#leaky relu
def leaky_relu(x):
  return max(0.1*x, x)
p = pd.Series(inp)
q = p.apply(leaky_relu)
```

```python
plt.plot(p, q, c='y')
```
