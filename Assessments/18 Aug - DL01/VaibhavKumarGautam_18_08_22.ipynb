```python
def tanh(x):
  th = ((math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x)))
  return th
```

```python
def relu(x):  
  return max(0.0,x)
```

```python
#leaky relu
```

```python
def leakyrelu(x):
  return max(0.1*x,x)
```

```python
leakyrelu(-100)
```

```python
def tanh(x):
  th = ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
  return th
```

```python
x = np.linspace(-10,10,30)
```

```python
plt.plot(x,tanh(x),c='r')
```

```python
def relu(x):
  data = [max(0,value) for value in x]
  return np.array(data, dtype=float)
```

```python
x_data = np.linspace(-10,10,100)
y_data = relu(x_data)
```

```python
plt.plot(x_data, y_data)
```
