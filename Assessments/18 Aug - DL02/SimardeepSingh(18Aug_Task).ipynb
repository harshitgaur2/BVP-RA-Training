```python
# Sigmoid (0,1)
import math 
def sigmoid(x):
  return 1.0/(1.0 + math.exp(-x))
```

```python
sigmoid(0.23)
```

```python
# tanh (-1,1)
def tanh(x):
  return (math.exp(x) - math.exp(-x))/(math.exp(x) + math.exp(-x))
```

```python
tanh(1)
```

```python
# Relu 
def relu(x):
  return max(0.0,x)
```

```python
relu(-100)
```

```python
relu(20)
```

```python
# Leaky Relu
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(-0.45)
```

```python
# Task - Plotting graphs of all the above activation functions 
```

```python
import matplotlib.pyplot as plt
import numpy as np
input = np.linspace(-15,15,200)
```

```python
#Sigmoid Graph
def sigmoid2(x):
  return 1.0/(1.0 + np.exp(-x))
plt.plot(input, sigmoid2(input), c="r")
```

```python
# Tanh Graph
def tanh2(x):
  return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
plt.plot(input, tanh2(input), c="g")
```

```python
# Relu Graph
def relu2(x):
  data = [max(0,value) for value in x]
  return np.array(data)
plt.plot(input, relu2(input), c="y")
```

```python
# Leaky Relu Graph
def leaky_relu2(x):
  data2 = [max(0.1*value,value) for value in x]
  return np.array(data2)
plt.plot(input, leaky_relu2(input), c="m")
```
