```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("ins_data.csv")

x=df.drop(columns="bought_insurance")
f=x.to_numpy()
label=df["bought_insurance"]
label=label.to_numpy()
label=label.reshape(28,1)
bias=np.random.rand(1)
df
```

```python
def sigmoid (x):
  return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):
    return sigmoid(x) * (1-sigmoid(x))
```

```python
w=np.random.rand(2,1) 
n=0.05
```

```python
for ep in range(10):
    inp=f
    sum=np.dot(f,w)+bias
    z=sigmoid(sum)
    error=z-label
    print(error.sum())
    dpred=error
    dz=sigmoid_der(z)
    delz=dpred*dz
    inp=sum.T
    w-=n*np.dot(inp,delz)
    
    for num in delz:
        bias-=n*num
```
