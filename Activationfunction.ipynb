```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input=np.linspace(-10,10,100)
```

**SIGMOID**



```python

def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
plt.plot(input,sigmoid(input))
```

```python
input=np.linspace(-10,10,100)

def tanh(x):
  return (np.exp(x)- np.exp(-x))/(np.exp(x) + np.exp(-x))
```

```python
plt.plot(input,tanh(input))
```

```python
def leakyrelu(x):
  x1=[]
  for i in x:
    if i>0 :
      x1.append(i)
    else :
      x1.append(0.01*i)
  return x1
```

```python
x = np.linspace(-5, 2, 100)
plt.plot(x, leakyrelu(x), 'r')
```
